#Numpy Array vs Python List

##Numpy array are faster, convinient and occupys less memory

```
list_a = range(100)
list_a
```

```
array_a = np.arange(100)
array_a
```

```
import sys
```

```
print(sys.getsizeof(87)*len(list_a))
```

```
print(array_a.itemsize*array_a.size)
```

from this we can see that numpy takes less space

```
import time
```

```
x=range(100000)
y=range(100000,200000)

#time function gives current time
start_time=time.time()

#It selects element from both array add them and puts the sum in another list
c=[(x+y) for x,y in zip(x,y)]
print(time.time()-start_time)
```

```
y=np.arange(100000)
x=np.arange(100000,200000)

#Now for performing same operation as above in numpy
start_time=time.time()
c=x+y

time.time()-start_time
```

from this we can see that numpy is faster then python

#Indexing with boolean arrays

```
#to get random integer in a range with given no of size
array12 = np.random.randint(low=1, high=10, size=15).reshape(5, 3)
array12
```

```
array12 > 5
```

```
array12 < 6
```

```
#to get array containing elements which are greater then 5, also know as indexing using boolean array
array12[array12 > 5]
```

```
array12[array12 < 6]
```

```
#to get elements which are greter then 5 and even
array12[((array12 > 5) & (array12%2 == 0))]
```

```
#to get elements which are grether then 5 or even
array12[((array12 > 5) | (array12%2 == 0))]
```
